```python

# Question 1
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def merge_lists(list1, list2):
    head = None
    tail = None

    while list1 is not None and list2 is not None:
        if list1.value >= list2.value:
            new_node = Node(list1.value)
            list1 = list1.next
        else:
            new_node = Node(list2.value)
            list2 = list2.next

        if head is None:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    while list1 is not None:
        new_node = Node(list1.value)
        list1 = list1.next

        if head is None:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    while list2 is not None:
        new_node = Node(list2.value)
        list2 = list2.next

        if head is None:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    return head
```

```python
# Question 2
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def remove_duplicates(head):
    if head is None:
        return head

    current = head

    while current.next is not None:
        if current.value == current.next.value:
            current.next = current.next.next
        else:
            current = current.next

    return head
```

```python
# Question 3
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def reverse_k_nodes(head, k):
    if head is None or k == 1:
        return head

    dummy = Node(0)
    dummy.next = head
    prev = dummy

    while True:
        count = 0
        current = prev.next
        while current:
            count += 1
            current = current.next
            if count == k:
                break

        if count < k:
            break

        current = prev.next
        next_node = None
        for _ in range(k):
            temp = current.next
            current.next = next_node
            next_node = current
            current = temp

        prev.next.next = current
        temp = prev.next
        prev.next = next_node
        prev = temp

    return dummy.next
```

```python
# Question 4
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def reverse_alternate_k_nodes(head, k):
    if head is None or k == 1:
        return head

    dummy = Node(0)
    dummy.next = head
    prev = dummy
    current = head
    count = 0
    while current:
        count += 1
        if count % (2 * k) == k + 1:
            start = current
        elif count % (2 * k) == 0:
            end = current
            next_node = None
            for _ in range(k):
                temp = start.next
                start.next = next_node
                next_node = start
                start = temp

            prev.next = next_node
            end.next = current

        prev = current
        current = current.next

    return dummy.next
```

```python
# Question 5
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def delete_last_occurrence(head, key):
    if head is None:
        return head

    prev = None
    last = None
    current = head

    while current:
        if current.value == key:
            last = current
        prev = current
        current = current.next

    if last is None:
        return head

    if last == head:
        head = head.next
    else:
        prev.next = last.next

    return head
```

```python
# Question 6
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def merge_sorted_lists(a, b):
    dummy = Node(0)
    merged = dummy

    while a and b:
        if a.value <= b.value:
            merged.next = a
            a = a.next
        else:
            merged.next = b
            b = b.next
        merged = merged.next

    if a:
        merged.next = a
    if b:
        merged.next = b

    return dummy.next
```

```python
# Question 7
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

def reverse_doubly_linked_list(head):
    if head is None or head.next is None:
        return head

    current = head
    prev = None

    while current:
        next_node = current.next
        current.next = prev
        current.prev = next_node
        prev = current
        current = next_node

    head = prev
    head.prev = None

    return head
```

```python
# Question 8
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

def delete_node_at_position(head, position):
    if position < 1 or head is None:
        return head

    if position == 1:
        new_head = head.next
        if new_head:
            new_head.prev = None
        return new_head

    current = head
    count = 1

    while current and count < position:
        current = current.next
        count += 1

    if current is None:
        return head

    prev_node = current.prev
    prev_node.next = current.next

    if current.next:
        current.next.prev = prev_node

    return head
```
